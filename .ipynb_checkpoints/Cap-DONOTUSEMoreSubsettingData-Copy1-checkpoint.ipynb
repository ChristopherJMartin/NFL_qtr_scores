### Trying to use neural network again

```python
games_Vscores_trainset = games_Vscores[games_Vscores['Year'] < 2016]
games_Vscores_trainset.tail()
```

```python
games_Vscores_testset = games_Vscores.query('Year >= 2016 & Year < 2019')
games_Vscores_testset.head()
```

```python
games_Vscores_trainbears = games_Vscores_trainset[games_Vscores_trainset.Vteam=='Bears'][['VQ1']]
games_Vscores_trainbears.T
```

```python
games_Vscores_testbears = games_Vscores_testset[games_Vscores_testset.Vteam=='Bears'][['VQ1']]
games_Vscores_testbears.T
```

```python
#set new variables
VBears = games_Vscores_bears.T
ValidBears = games_Vscores_test_bears.T
```

```python
ValidBears
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X = games_Vscores_trainbears
y = games_Vscores_testbears
```

```python
y
```

```python
model=Sequential()
model.add(Dense(40, activation = 'relu'))
model.add(Dense(8, activation = 'softmax'))
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['acc'])
```

```python
history = model.fit(X,y,
                    validation_data = (games_Vscores_test_bears),
                   epochs=20)
```
