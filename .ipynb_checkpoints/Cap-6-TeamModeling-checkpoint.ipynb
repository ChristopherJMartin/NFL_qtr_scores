```python
# Visiting team: take quarter winner codes, one hot encode, encode winner column as well
Vteams_dict = {}
for team in teamlist1:
    print(team, type(team))
    #teamfile = team + '_Vteam_quarterly_scores'
    team_Vteam_Q_wins = qwins[qwins['Vteam'] == team]
    team_Vteam_quarterly_wins = team_Vteam_Q_wins[['RQ1','RQ2','RQ3','RQ4']]
    ohe1 = OneHotEncoder()
    team_Vteam_quarterly_wins_encoded = pd.DataFrame(ohe1.fit_transform(team_Vteam_quarterly_wins).toarray(),index=team_Vteam_quarterly_wins.index)
    ohe1.get_feature_names()
    team_Vteam_quarterly_wins_encoded.columns = ohe1.get_feature_names()
    team_Vteam_quarterly_wins_for_modeling = pd.concat([team_Vteam_quarterly_wins_encoded, team_Vteam_Q_wins['Vteam'], team_Vteam_Q_wins['Winner']],axis=1)
    winner_codes = {"Winner":     {"HW": -1, "VW": 1, "T": 0}}
    team_Vteam_quarterly_wins_for_modeling = team_Vteam_quarterly_wins_for_modeling.replace(winner_codes)
    Vteams_dict[team] = team_Vteam_quarterly_wins_for_modeling
```

```python
# make list of qtr clumps
clumps = [Clump_Q1,Clump_Q2,Clump_Q3,Clump_Q4,Clump_Q1_Q2,Clump_Q2_Q3,Clump_Q3_Q4,
          Clump_Q1_Q3,Clump_Q1_Q4,Clump_Q2_Q4,Clump_Q1_Q2_Q3,Clump_Q1_Q3_Q4,Clump_Q1_Q2_Q4,
         Clump_Q2_Q3_Q4,Clump_Q1_Q2_Q3_Q4]
```

```python
# Home team: take quarter winner codes, one hot encode, encode winner column as well (move this cell up?)
Hteams_dict = {}
for team in teamlist1:
    print(team, type(team))
    #teamfile = team + '_Vteam_quarterly_scores'
    team_Hteam_Q_wins = qwins[qwins['Hteam'] == team]
    team_Hteam_quarterly_wins = team_Hteam_Q_wins[['RQ1','RQ2','RQ3','RQ4']]
    ohe1 = OneHotEncoder()
    team_Hteam_quarterly_wins_encoded = pd.DataFrame(ohe1.fit_transform(team_Hteam_quarterly_wins).toarray(),index=team_Hteam_quarterly_wins.index)
    ohe1.get_feature_names()
    team_Hteam_quarterly_wins_encoded.columns = ohe1.get_feature_names()
    team_Hteam_quarterly_wins_for_modeling = pd.concat([team_Hteam_quarterly_wins_encoded, team_Hteam_Q_wins['Hteam'], team_Hteam_Q_wins['Winner']],axis=1)
    winner_codes = {"Winner":     {"HW": -1, "VW": 1, "T": 0}}
    team_Hteam_quarterly_wins_for_modeling = team_Hteam_quarterly_wins_for_modeling.replace(winner_codes)
    Hteams_dict[team] = team_Hteam_quarterly_wins_for_modeling
    
```

```python
### trying loop for estimators - home team
dict_classifiers = {
    "Logr": LogisticRegression(),
    "KNN": KNeighborsClassifier(),
    "bags": BaggingClassifier(),
    "forest": RandomForestClassifier(), 
    "Xtrees": ExtraTreesClassifier(),
    "DT": DecisionTreeClassifier(),
    "AB": AdaBoostClassifier(),
    "Bayes": MultinomialNB(),
    "GBC": GradientBoostingClassifier()
}
record = pd.DataFrame(columns=['Team','Clump', 'Estimator', 'params', 'accuracy'])
for indx, Hteam in enumerate(teamlist1):
    Hteam_wins = Hteams_dict[Hteam]
    for index, clump in enumerate(clumps):
        try:
            X = Hteam_wins[clump]
        except:
            continue
        y = Hteam_wins['Winner']
        X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 32221)
        for model, model_instantiation in dict_classifiers.items():
            model = model_instantiation # change estimator each time
        # and GridSearch here?
            model.fit(X_train, y_train)
            pred = model.predict(X_test)
            score = accuracy_score(y_test, pred)
            print(score)
            record = record.append({'Team': Hteam, 'Clump': index, 'Estimator': model_instantiation, 
                       'params': 'NA', 'accuracy': score}, ignore_index=True)
```

```python
### trying loop for estimators - visiting team
dict_classifiers = {
    "Logr": LogisticRegression(),
    "KNN": KNeighborsClassifier(),
    "bags": BaggingClassifier(),
    "forest": RandomForestClassifier(), 
    "Xtrees": ExtraTreesClassifier(),
    "DT": DecisionTreeClassifier(),
    "AB": AdaBoostClassifier(),
    "Bayes": MultinomialNB(),
    "GBC": GradientBoostingClassifier()
}
for indx, Vteam in enumerate(teamlist1):
    Vteam_wins = Vteams_dict[Vteam]
    for index, clump in enumerate(clumps):
        try:
            X = Vteam_wins[clump]
        except:
            continue
        y = Vteam_wins['Winner']
        X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 32221)
        for model, model_instantiation in dict_classifiers.items():
            model = model_instantiation 
            model.fit(X_train, y_train)
            pred = model.predict(X_test)
            score = accuracy_score(y_test, pred)
            print(score)
            record = record.append({'Team': Vteam, 'Clump': index, 'Estimator': model_instantiation, 
                       'params': 'NA', 'accuracy': score}, ignore_index=True)
```
