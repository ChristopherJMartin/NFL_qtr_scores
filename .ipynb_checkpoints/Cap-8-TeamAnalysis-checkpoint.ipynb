```python
def team_mean(team, HV):
    '''
    Enter team name (and home or visitor), get back mean accuracy by clump
    '''
    oneteam = teams[(teams['Team'] == team) & (teams['HV'] == HV)]
    oneteam_subset = oneteam[['Clump','accuracy']]
    result = oneteam_subset.groupby('Clump').accuracy.mean()
    print(f'Here are the mean accuracy scores by clump for the {team}:')
    return result
```

```python
## function
def clump_max_acc(clump, HV):
    '''
    Enter clump number (and home or visitor), return maximum accuracy by team
    '''
    oneclump = teams[(teams['Clump'] == clump) & (teams['HV'] == HV)]
    oneteam_subset = oneteam[['Clump','accuracy']]
    result = oneteam_subset.groupby('Clump').accuracy.mean()
    print(f'Here are the mean accuracy scores by clump for the {HV} team:')
    return result
```

```python
#columns=['Team','Estimator','accuracy']
team_max_by_clump_list = []
for clump in range(0,15):
    print(f'For Clump {clump}:')
    for team in teamlist:
        old_acc = 0
        est = 'X'
        for teamindex in range(0,teams.shape[0]):
            if (team == teams.iloc[teamindex,0]) & (clump == teams.iloc[teamindex,1]):
                new_acc = teams.iloc[teamindex,4]
                if new_acc > old_acc:
                    old_acc = new_acc
                    est = teams.iloc[teamindex,2]
                    HV = teams.iloc[teamindex,5]
        print(team,est,old_acc,HV)
        team_max_by_clump_list.append([clump, team,est,old_acc,HV])
        #temp = pd.DataFrame({'Team': team, 'Estimator': est, 'accuracy': old_acc}, index=oneclump_subset.index)
        #team_max_by_clump = pd.concat(team_max_by_clump, temp)
            
```

```python
team_max_by_clump = pd.DataFrame(team_max_by_clump_list, columns=['Clump','Team','Estimator','accuracy', 'HV'])
team_max_by_clump
```

```python
team_min_by_clump_list = []
for clump in range(0,15):
    print(f'For Clump {clump}:')
    for team in teamlist:
        old_acc = 1
        est = 'X'
        for teamindex in range(0,oneclump_subset.shape[0]):
            if team == oneclump_subset.iloc[teamindex,0]:
                new_acc = oneclump_subset.iloc[teamindex,2]
                if new_acc < old_acc:
                    old_acc = new_acc
                    est = oneclump_subset.iloc[teamindex,1]
                    HV = teams.iloc[teamindex,5]
        print(team,est,old_acc)
        team_min_by_clump_list.append([clump, team,est,old_acc, HV])
```

```python
team_min_by_clump = pd.DataFrame(team_min_by_clump_list, columns=['Clump','Team','Estimator','accuracy','HV'])
team_min_by_clump
```
