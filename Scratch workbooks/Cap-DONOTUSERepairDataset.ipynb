### Must Repair Week column in data

```python
games = games.drop(['Unnamed: 0'], axis=1)
games
```

```python
import re
for i in range(1,games.shape[0]):
    if games.iloc[i,10] > 2015 & games.iloc[i,10] < 2019:
        games.iloc[i,2] = re.search('-(.*)-',games.iloc[i,1]).group(1)
        
```

```python
for game in range(1,5324):
    value = games.iloc[game,2]
    if value == 'WC':
        games.iloc[game,2] = 100
    elif value == 'DV':
        games.iloc[game,2] = 200
    elif value == 'CC':
        games.iloc[game,2] = 300
    elif value == 'SB':
        games.iloc[game,2] = 400
```
