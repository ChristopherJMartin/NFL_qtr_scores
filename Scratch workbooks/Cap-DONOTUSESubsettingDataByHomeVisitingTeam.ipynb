### One Hot Encode Scores https://stackoverflow.com/questions/37292872/how-can-i-one-hot-encode-in-python

```python
one_hot_V = pd.get_dummies(games_Vscores['Vteam'])
```

```python
games_Vscores_enc = games_Vscores.drop(['Vteam'], axis=1)
```

```python
games_Vscores_enc = games_Vscores_enc.join(one_hot_V)
```

```python
games_Vscores_enc
```

```python
games_Vscores.to_csv('data/NFL_games_Vscores_teams_encoded.csv')
```

```python
one_hot_H = pd.get_dummies(games_Hscores['Hteam'])
```

```python
games_Hscores_enc = games_Hscores.drop(['Hteam'], axis=1)
```

```python
games_Hscores_enc = games_Hscores_enc.join(one_hot_H)
```
