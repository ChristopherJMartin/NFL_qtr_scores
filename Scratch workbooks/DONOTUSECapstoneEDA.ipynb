```python
for game in range(1,5324):
    value = games.iloc[game,1]
    if value == 'WildCard':
        games.iloc[game,1] = 100
    elif value == 'Division':
        games.iloc[game,1] = 200
    elif value == 'ConfChamp':
        games.iloc[game,1] = 300
    elif value == 'SuperBowl':
        games.iloc[game,1] = 400
```

```python
games['Week'] = pd.to_numeric(games['Week'])
```
