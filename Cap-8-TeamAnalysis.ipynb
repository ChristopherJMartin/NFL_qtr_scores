```python
## function
def team_mean(team, HV):
    '''
    Enter team name (and home or visitor), return mean accuracy by clump
    '''
    oneteam = teams[(teams['Team'] == team) & (teams['HV'] == HV)]
    oneteam_subset = oneteam[['Clump','accuracy']]
    result = oneteam_subset.groupby('Clump').accuracy.mean()
    print(f'Here are the mean accuracy scores by clump for the {team} as the {HV} team:')
    return result
```

```python
teamlist = teams['Team'].unique()
teamlist
```

```python
#Max accuracy for each team by clump
team_max_by_clump_list = []
for clump in range(0,15):
    print(f'For Clump {clump}:')
    for team in teamlist:
        old_acc = 0
        est = 'X'
        for teamindex in range(0,teams.shape[0]):
            if (team == teams.iloc[teamindex,0]) & (clump == teams.iloc[teamindex,1]):
                new_acc = teams.iloc[teamindex,4]
                if new_acc > old_acc:
                    old_acc = new_acc
                    est = teams.iloc[teamindex,2]
                    HV = teams.iloc[teamindex,5]
        print(team,est,old_acc,HV)
        team_max_by_clump_list.append([clump, team,est,old_acc,HV])
            
```

```python
team_max_by_clump = pd.DataFrame(team_max_by_clump_list, columns=['Clump','Team','Estimator','accuracy', 'HV'])
team_max_by_clump
```

```python
#Min accuracy for each team by clump
team_min_by_clump_list = []
for clump in range(0,15):
    print(f'For Clump {clump}:')
    for team in teamlist:
        old_acc = 1
        est = 'X'
        for teamindex in range(0,teams.shape[0]):
            if team == teams.iloc[teamindex,0]:
                new_acc = teams.iloc[teamindex,4]
                if new_acc < old_acc:
                    old_acc = new_acc
                    est = teams.iloc[teamindex,1]
                    HV = teams.iloc[teamindex,5]
        print(team,est,old_acc)
        team_min_by_clump_list.append([clump, team,est,old_acc, HV])
```

```python
team_min_by_clump = pd.DataFrame(team_min_by_clump_list, columns=['Clump','Team','Estimator','accuracy','HV'])
team_min_by_clump
```
